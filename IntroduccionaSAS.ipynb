  Podemos tener tres tipos de archivos
 - Archivos de datos (raw file), archivos de texto, de excel, en formato de coma csv
 - Archivos de datos SAS (SAS data set), los datos de sas se guardan en este tipo de archivos para su lectura o escritura
 - Archivos de programas SAS, tal cual programas escritos en SAS
 
 Un programa de SAS solo contiene
 - Data. Por lo regular tiene como entrada un conjunto de datos (raw file) y produce un SAS data set. Las variables son las columnas del sas data set, y pueden generarse nuevas columnas a partir de las que tenia el raw file. 
 - Proc. De forma tipica trabaja con un data set y gebera reportes y gráficas, 
 - Una combinación de los anteriores
 
 Cada programa tiene bloques tipo data y bloques tipo proc, es recomendable terminar cada bloque con un run, cada intrucción (o sentencia) lleva al final un punto y coma. Las instrucciones que estan fuera de los bloques reciben el nombre de instrucciones globales
 
 <code>
 data ......;
 sentencia_a .....;
 sentencia_b.....;
 .
 .
 sentecnia_n
 run;
 </code>
 <code>
 sentencia;
 </code>
 <code>
proc ......;
 sentencia_1 .....;
 sentencia_2.....;
 .
 .
 sentecnia_m
run;
 </code>
 
 La sintaxys es libre no hay diferencia entre mayusculas o minúsculas lo cual podría ser bueno o malo, también puede comensar una instrucción donde se quiera incluso todo el programa podría estra en una sola línea, pero existe una guia de buenas constumbres, en SAS Studio el botón format code hace esto. 
 
 Se puedes escribir comentarios de diferencites tipos, el comentarios de bloque por ejemplo 
 
 <code>
 / ******* cometario esto *****/
 
 * Comentario qeu ocmienzan con asterisco tienene que terminar con ; esto permite comentar instrucciones
 </code>
